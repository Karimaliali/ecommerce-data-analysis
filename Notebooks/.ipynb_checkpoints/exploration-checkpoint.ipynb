```python
import pandas as pd
import numpy as np
data = pd.read_csv(
    r'C:\Users\LORD LAPTOP\Documents\GitHub\ecommerce-data-analysis\data\raw\digital_market.csv',
    encoding='latin1'
)
data.head()
```

```python
data.columns
data.info()
print(data.describe())
print(data.isnull().sum())
```

```python
data['TotalPrice'] = data['Quantity']*data['UnitPrice']
data.head()
```

```python
data =  data[(data['Quantity']>0)&(data['UnitPrice']>0)]
data.shape
```

```python
data = data.dropna(subset=['CustomerID'])
data.shape
```

```python
data['InvoiceDate'] = pd.to_datetime(data['InvoiceDate'])
data.info
```

```python
top_countries = data['Country'].value_counts().head(10)
top_countries
```

```python
import matplotlib.pyplot as plt 
top10_countries = top_countries.head()
plt.figure(figsize=(10,6))
top10_countries.plot(kind='bar', color = 'skyblue')
plt.title('Top 10 Countries By Number Of Orders', fontsize =14)
plt.xlabel('Country')
plt.ylabel('Number of Orders')
plt.xticks(rotation=45)
plt.show()
```

```python
top_products = data['Description'].value_counts().head(10)
top_products
```

```python
top_revenue_products = data.groupby('Description')['TotalPrice'].sum().sort_values(ascending=False).head(10)
top_revenue_products
```

```python
data['Month'] = data['InvoiceDate'].dt.to_period('M')
```

```python
monthly_sales = data.groupby('Month')['TotalPrice'].sum()
monthly_sales
```

```python
import matplotlib.pyplot as plt
monthly_sales = data.groupby('Month')['TotalPrice'].sum()
plt.figure(figsize=(10,6))
plt.plot(monthly_sales.index.astype(str),monthly_sales.values, marker = 'o',linewidth=2)
plt.title('Monthly Sales', fontsize=16)
plt.xlabel('Month')
plt.ylabel('Total Sales ()')
plt.grid(True)
plt.show()
```

```python
top_customers = data.groupby('CustomerID')['TotalPrice'].sum().sort_values(ascending=False).head(10)
top_customers
```

```python
import matplotlib.pyplot as plt
top_customers.plot(kind='bar',figsize=(10,5),color ='orange')
plt.title('Total 10 Customer by Total Spending')
plt.xlabel('Customer ID')
plt.ylabel('Total Spending ()')
plt.show()
```

```python
top_products = data.groupby('Description')['Quantity'].sum().sort_values(ascending=False).head(10)
top_products.plot(kind ='bar',figsize=(10,5),color='teal')
plt.title('Top 10 Best-Selling Products')
plt.xlabel('Product')
plt.ylabel('Total Quantity Sold')
plt.show()
```

```python
monthly_sales = data.groupby('Month')['TotalPrice'].sum()
plt.figure(figsize=(10,5))
monthly_sales.plot(kind='line',marker='o', color='green')
plt.title('Monthly Sales Trend')
plt.xlabel('Month')
plt.ylabel('Total Sales ()')
plt.grid(True)
plt.show()
```

```python
plt.figure(figsize=(14,8))
plt.scatter(data['UnitPrice'],data['Quantity'],alpha=0.2,color='purple')
plt.title('Relationship Between Unit Price and Quantity Sold')
plt.xlabel('Unit Price ()')
plt.ylabel('Quantity Sold')
plt.xscale('log')
plt.yscale('log')
plt.grid(True)
plt.show()
```
